```
import random
import math
```

```
x_i = []
y_o = []
for i in range(1,31):
  x_i.append(i/4)
random.shuffle(x_i)
for i in range(0,30):
  y_o.append(x_i[i]**(4/5))
x_inFull = [((x_i[k]-min(x_i))/(max(x_i)-min(x_i))) for k in range(len(x_i))]
#x_inFull=x_i
print(x_inFull)
y_outFull = y_o
print(y_outFull)
```

```
from random import randrange
from random import random
from math import exp
```

```
def network(n_hidden):
  network=[]
  hidden=[random() for i in range(n_hidden)]
  final=[random() for i in range(n_hidden)]
  network.append(hidden)
  network.append(final)
  bias = []
  b1 = [random() for i in range(n_hidden)]
  b2 = random() 
  bias.append(b1)
  bias.append(b2)
  return network, bias
```

```
def sigmoid(activation):
	return 1.0 / (1.0 + exp(-activation))
```

```
def forward(network,bias,x):
  first = network[0]
  second = network[1]
  b1 = bias[0]
  b2 = bias[1]
  zo = []
  zi = []
  for i in range(len(first)):
    z = b1[i] + first[i]*x
    zi.append(z)
    z = sigmoid(z)
    zo.append(z)
  y = b2
  for i in range(len(second)):
    y = y + zo[i]*second[i]
  return y,zo,zi
```

```
def derivative(output):
	return output * (1.0 - output)
```

```
def backprop(network, bias, error,zo,zi,x,a,m):
  dk = error
  for i in range(len(network[1])):
    network[1][i] = network[1][i] + m*a*dk*zo[i] #weight updation no activation at output
  for i in range(len(network[0])):
    network[0][i] = network[0][i] + m*a*dk*network[1][i]*derivative(zo[i])*x #weight updation.. activation is there
  for i in range(len(bias[0])):
    bias[0][i] = bias[0][i] + m*a*dk*network[1][i]*derivative(zo[i])
  bias[1] = bias[1] + m*a*dk
```

```
Network,bias = network(5)
x_in = x_inFull[0:24]
y_out = y_outFull[0:24]
```

```
epochs = 10000
a=0.01
m=1
for epoch in range(1,epochs+1):
  error_tot = 0
  for i in range(len(x_in)):
    x = x_in[i]
    y,zo,zi = forward(Network,bias, x)
    y = y*(max(y_outFull)-min(y_outFull)) + min(y_outFull)
    error = y_out[i] - y
    backprop(Network,bias, error,zo,zi,x,a,m)
    error_tot = error_tot + (error**2)/2
  print('>epoch=%d, error=%.3f' % (epoch,error_tot), )
```

```
def predict(Network,bias, x):
  
  y,zo,zi = forward(Network,bias,x)
  y = y*(max(y_out)-min(y_out)) + min(y_out)
  return y
```

```
x_test = x_inFull[24:30]
y_test = y_outFull[24:30]
```

```
Accuracy = 0
for i in range(len(x_test)):
  prediction = predict(Network,bias,x_test[i])
  Accuracy = Accuracy + 100 - abs(y_test[i] - prediction)/prediction*100
  print('Expected=%.4f, Got=%.4f' % (y_test[i], prediction),)
Accuracy/=len(x_test)
print('Accuracy = '+str(Accuracy))
```
