```
import random
```

```
x_i = []
y_o = []
for i in range(1,31):
  x_i.append(i)
random.shuffle(x_i) # Total ip data set which is a random order of the numbers from 1 to 31
for i in range(0,30):
  y_o.append(x_i[i]**(4/5)) # Total op data set for the given ip data for the funtion y=x^(4/5)
x_inFull = [((x_i[k]-min(x_i))/(max(x_i)-min(x_i))) for k in range(len(x_i))] # Normalizing the input w.r.t to the dataset to obtain better performance
print(x_inFull)
y_outFull = y_o # We don't normalize the output 
print(y_outFull)
```

```
from random import randrange
from random import random
from math import exp
```

```
#defining the network with one hidden layer and n_hidden number of neurons
def network(n_hidden):
  network=[]
  hidden=[random() for i in range(n_hidden)] #weights between ip and hidden layer
  final=[random() for i in range(n_hidden)] #weights between hidden layer and op
  network.append(hidden)
  network.append(final)
  bias = []
  b1 = [random() for i in range(n_hidden)] #bias for hidden layer
  b2 = random() #bias for output
  bias.append(b1)
  bias.append(b2)
  return network, bias
```

```
# The activation function that we will be usinng is sigmoid because of it's easy differentability
def sigmoid(activation):
	return 1.0 / (1.0 + exp(-activation))
```

```
# Feed forward for a particular ip x
def forward(network,bias,x):
  first = network[0]
  second = network[1]
  b1 = bias[0]
  b2 = bias[1]
  zo = [] # op of hidden layer
  zi = [] #ip of hidden layer
  for i in range(len(first)):
    z = b1[i] + first[i]*x
    zi.append(z)
    z = sigmoid(z)
    zo.append(z)
  y = b2
  for i in range(len(second)):
    y = y + zo[i]*second[i]
  return y,zo,zi
```

```
def derivative(output):
	return output * (1.0 - output)
```

```
def backprop(network, bias, error,zo,zi,x,a,m):
  dk = error*1 #derivative of y=x is 1
  for i in range(len(network[1])):
    network[1][i] = network[1][i] + m*a*dk*zo[i] #weight updation no activation at output
  for i in range(len(network[0])):
    network[0][i] = network[0][i] + m*a*dk*network[1][i]*derivative(zo[i])*x #weight updation.. activation is there
  for i in range(len(bias[0])):
    bias[0][i] = bias[0][i] + m*a*dk*network[1][i]*derivative(zo[i]) #bias updation
  bias[1] = bias[1] + m*a*dk
```

```
Network,bias = network(5) # 5 hidden neurons
x_in = x_inFull[0:24] #24 data pts for training and rest for validation
y_out = y_outFull[0:24]
```

```
epochs = 5000
a=0.01 #learning rate
m=1    #momentum
for epoch in range(1,epochs+1):
  error_tot = 0
  for i in range(len(x_in)):
    x = x_in[i]
    y,zo,zi = forward(Network,bias, x)
    y = y*(max(y_outFull)-min(y_outFull)) + min(y_outFull) #denormalization of the op from the network
    error = y_out[i] - y
    backprop(Network,bias, error,zo,zi,x,a,m)
    error_tot = error_tot + (error**2)/2
  print('>epoch=%d, error=%.3f' % (epoch,error_tot), )
```

```
# Function to calculate the op for the testing part
# The y op is a normalized op. So, we are de normalizing it before returning it
def predict(Network,bias, x):
  
  y,zo,zi = forward(Network,bias,x)
  y = y*(max(y_out)-min(y_out)) + min(y_out)
  return y
```

```
# 6 data points for testing
x_test = x_inFull[24:30]
y_test = y_outFull[24:30]
```

```
# Testing phase
Accuracy = 0
for i in range(len(x_test)):
  prediction = predict(Network,bias,x_test[i])
  Accuracy = Accuracy + 100 - abs(y_test[i] - prediction)/prediction*100 # average accuracy
  print('Expected=%.4f, Got=%.4f' % (y_test[i], prediction),)
Accuracy/=len(x_test)
print('Accuracy = '+str(Accuracy))
```
